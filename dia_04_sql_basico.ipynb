```python
from pathlib import Path
import sqlite3
import pandas as pd

# Carpeta del Día 4.
CARPETA_DATOS = Path("datos_dia_04")
CARPETA_DATOS.mkdir(exist_ok=True)

# Base de datos que se usará durante esta sesión.
RUTA_DB = CARPETA_DATOS / "ventas_dia_04.db"

print("Carpeta de trabajo:")
print(CARPETA_DATOS.resolve())

print("\nBase de datos del Día 4:")
print(RUTA_DB.resolve())
```

```python
conexion = sqlite3.connect(RUTA_DB)

print("Conexión creada correctamente.")
print("Si la base no existía, SQLite creó el archivo ventas_dia_04.db.")

conexion.close()

print("Conexión cerrada correctamente.")
```

```python
conexion = sqlite3.connect(RUTA_DB)
cursor = conexion.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS productos (
    id_producto INTEGER PRIMARY KEY AUTOINCREMENT,
    sku TEXT NOT NULL UNIQUE,
    nombre TEXT NOT NULL,
    precio REAL NOT NULL CHECK(precio > 0),
    stock INTEGER NOT NULL DEFAULT 0 CHECK(stock >= 0),
    categoria TEXT DEFAULT 'General'
)
""")

conexion.commit()
conexion.close()

print("Tabla productos creada correctamente.")
```

```python
def intentar_insertar_producto(sku, nombre, precio, stock, categoria=None):
    conexion = sqlite3.connect(RUTA_DB)
    cursor = conexion.cursor()

    try:
        cursor.execute("""
        INSERT INTO productos (sku, nombre, precio, stock, categoria)
        VALUES (?, ?, ?, ?, ?)
        """, (sku, nombre, precio, stock, categoria))

        conexion.commit()
        print(f"Producto insertado correctamente: {sku} - {nombre}")

    except sqlite3.IntegrityError as error:
        print("No se pudo insertar el producto.")
        print("Motivo:", error)

    finally:
        conexion.close()
```

```python
from pathlib import Path
import sqlite3
import pandas as pd

CARPETA_DATOS = Path("datos_dia_04")
CARPETA_DATOS.mkdir(exist_ok=True)

RUTA_DB = CARPETA_DATOS / "ventas_dia_04.db"

conexion = sqlite3.connect(RUTA_DB)
cursor = conexion.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS productos (
    id_producto INTEGER PRIMARY KEY AUTOINCREMENT,
    sku TEXT NOT NULL UNIQUE,
    nombre TEXT NOT NULL,
    precio REAL NOT NULL CHECK(precio > 0),
    stock INTEGER NOT NULL DEFAULT 0 CHECK(stock >= 0),
    categoria TEXT DEFAULT 'General'
)
""")

productos = [
    ("P001", "Mouse inalámbrico", 249.90, 15, "Accesorios"),
    ("P002", "Teclado mecánico", 899.00, 8, "Accesorios"),
    ("P003", "Monitor 24 pulgadas", 3299.00, 4, "Pantallas"),
    ("P004", "Cable HDMI", 120.00, 20, "Cables"),
    ("P005", "Memoria USB 64GB", 150.00, 30, "Almacenamiento")
]

for producto in productos:
    try:
        cursor.execute("""
        INSERT INTO productos (sku, nombre, precio, stock, categoria)
        VALUES (?, ?, ?, ?, ?)
        """, producto)
    except sqlite3.IntegrityError:
        pass

conexion.commit()

consulta = pd.read_sql_query("""
SELECT
    id_producto,
    sku,
    nombre,
    precio,
    stock,
    categoria,
    precio * stock AS valor_inventario
FROM productos
ORDER BY precio DESC
""", conexion)

conexion.close()

consulta
```
