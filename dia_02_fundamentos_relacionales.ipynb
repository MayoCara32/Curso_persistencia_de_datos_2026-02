```python
def validar_llave_primaria(tabla, campo_pk):
    valores = tabla[campo_pk]

    cantidad_total = len(valores)
    cantidad_unicos = valores.nunique()

    if cantidad_total == cantidad_unicos:
        print(f"El campo '{campo_pk}' puede funcionar como llave primaria.")
    else:
        print(f"El campo '{campo_pk}' NO puede funcionar como llave primaria.")
        print("Hay valores repetidos.")

validar_llave_primaria(tabla_productos, "id_producto")
```

```python
clientes = [
    {
        "id_cliente": 1,
        "nombre": "Ana López",
        "correo": "ana@example.com"
    },
    {
        "id_cliente": 2,
        "nombre": "Carlos Pérez",
        "correo": "carlos@example.com"
    },
    {
        "id_cliente": 3,
        "nombre": "María Torres",
        "correo": "maria@example.com"
    }
]

ventas = [
    {
        "id_venta": 1,
        "id_cliente": 1,
        "fecha": "2026-06-09",
        "total": 499.80
    },
    {
        "id_venta": 2,
        "id_cliente": 2,
        "fecha": "2026-06-09",
        "total": 899.00
    },
    {
        "id_venta": 3,
        "id_cliente": 1,
        "fecha": "2026-06-10",
        "total": 3299.00
    }
]

tabla_clientes = pd.DataFrame(clientes)
tabla_ventas = pd.DataFrame(ventas)

print("Tabla clientes:")
display(tabla_clientes)

print("Tabla ventas:")
display(tabla_ventas)
```

```python
def validar_llave_foranea(tabla_hija, campo_fk, tabla_padre, campo_pk):
    valores_fk = set(tabla_hija[campo_fk])
    valores_pk = set(tabla_padre[campo_pk])

    valores_invalidos = valores_fk - valores_pk

    if not valores_invalidos:
        print(f"La llave foránea '{campo_fk}' es válida.")
        print("Todas las ventas apuntan a clientes existentes.")
    else:
        print(f"La llave foránea '{campo_fk}' tiene valores inválidos.")
        print("Estos valores no existen en la tabla padre:")
        print(valores_invalidos)

validar_llave_foranea(
    tabla_hija=tabla_ventas,
    campo_fk="id_cliente",
    tabla_padre=tabla_clientes,
    campo_pk="id_cliente"
)
```

```python
import pandas as pd

clientes = pd.DataFrame([
    {"id_cliente": 1, "nombre": "Ana López", "correo": "ana@example.com"},
    {"id_cliente": 2, "nombre": "Carlos Pérez", "correo": "carlos@example.com"}
])

productos = pd.DataFrame([
    {"id_producto": 1, "nombre": "Mouse inalámbrico", "precio": 249.90, "stock": 15},
    {"id_producto": 2, "nombre": "Teclado mecánico", "precio": 899.00, "stock": 8},
    {"id_producto": 3, "nombre": "Monitor 24 pulgadas", "precio": 3299.00, "stock": 4}
])

ventas = pd.DataFrame([
    {"id_venta": 1, "id_cliente": 1, "fecha": "2026-06-09"},
    {"id_venta": 2, "id_cliente": 2, "fecha": "2026-06-10"}
])

detalle_ventas = pd.DataFrame([
    {"id_detalle": 1, "id_venta": 1, "id_producto": 1, "cantidad": 2, "precio_unitario": 249.90},
    {"id_detalle": 2, "id_venta": 1, "id_producto": 2, "cantidad": 1, "precio_unitario": 899.00},
    {"id_detalle": 3, "id_venta": 2, "id_producto": 3, "cantidad": 1, "precio_unitario": 3299.00}
])

detalle_ventas["subtotal"] = detalle_ventas["cantidad"] * detalle_ventas["precio_unitario"]

ventas_con_cliente = ventas.merge(clientes, on="id_cliente", how="left")

detalle_con_producto = detalle_ventas.merge(productos, on="id_producto", how="left")

total_por_venta = detalle_ventas.groupby("id_venta")["subtotal"].sum().reset_index()
total_por_venta = total_por_venta.rename(columns={"subtotal": "total"})

resumen_ventas = ventas_con_cliente.merge(total_por_venta, on="id_venta", how="left")

print("Clientes")
display(clientes)

print("Productos")
display(productos)

print("Ventas")
display(ventas)

print("Detalle de ventas")
display(detalle_ventas)

print("Ventas con nombre de cliente")
display(resumen_ventas[["id_venta", "fecha", "nombre", "correo", "total"]])

print("Detalle con nombre de producto")
display(detalle_con_producto[[
    "id_venta",
    "nombre",
    "cantidad",
    "precio_unitario",
    "subtotal"
]])
```
