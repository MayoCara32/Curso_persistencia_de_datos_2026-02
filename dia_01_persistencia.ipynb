```python
productos_memoria = []

def agregar_producto_memoria(nombre, precio, stock):
    producto = {
        "nombre": nombre,
        "precio": precio,
        "stock": stock
    }

    productos_memoria.append(producto)

def mostrar_productos_memoria():
    if not productos_memoria:
        print("No hay productos registrados en memoria.")
        return

    print("Productos registrados en memoria:")
    for producto in productos_memoria:
        print(f"{producto['nombre']} | ${producto['precio']} | Stock: {producto['stock']}")

agregar_producto_memoria("Mouse inalámbrico", 249.90, 15)
agregar_producto_memoria("Teclado mecánico", 899.00, 8)
agregar_producto_memoria("Monitor 24 pulgadas", 3299.00, 4)

mostrar_productos_memoria()
```

```python
print("Antes de borrar la variable:")
mostrar_productos_memoria()

# Esta línea simula que el programa terminó y la información en memoria se perdió.
del productos_memoria

print("\nDespués de borrar la variable:")

try:
    mostrar_productos_memoria()
except NameError:
    print("La información ya no existe porque estaba guardada solo en memoria.")
```

```python
from pathlib import Path
import csv
import json
from datetime import datetime

# Carpeta donde se guardarán los archivos generados durante la sesión.
CARPETA_DATOS = Path("datos_dia_01")
CARPETA_DATOS.mkdir(exist_ok=True)

print("Carpeta de trabajo lista:")
print(CARPETA_DATOS.resolve())

archivo_txt = CARPETA_DATOS / "ventas.txt"

def registrar_venta_txt(cliente, producto, cantidad, precio_unitario):
    total = cantidad * precio_unitario
    fecha = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    linea = f"{fecha} | Cliente: {cliente} | Producto: {producto} | Cantidad: {cantidad} | Total: ${total:.2f}\n"

    # Modo "a": agrega la venta al final del archivo sin borrar lo anterior.
    with open(archivo_txt, "a", encoding="utf-8") as archivo:
        archivo.write(linea)

    print("Venta registrada en archivo TXT.")

def mostrar_ventas_txt():
    if not archivo_txt.exists():
        print("Todavía no existe el archivo de ventas.")
        return

    with open(archivo_txt, "r", encoding="utf-8") as archivo:
        contenido = archivo.read()

    print("Ventas guardadas en TXT:")
    print(contenido)

registrar_venta_txt("Ana López", "Mouse inalámbrico", 2, 249.90)
registrar_venta_txt("Carlos Pérez", "Teclado mecánico", 1, 899.00)

mostrar_ventas_txt()
```

```python
archivo_csv = CARPETA_DATOS / "productos.csv"

productos = [
    {"sku": "P001", "nombre": "Mouse inalámbrico", "precio": 249.90, "stock": 15},
    {"sku": "P002", "nombre": "Teclado mecánico", "precio": 899.00, "stock": 8},
    {"sku": "P003", "nombre": "Monitor 24 pulgadas", "precio": 3299.00, "stock": 4}
]

with open(archivo_csv, "w", encoding="utf-8", newline="") as archivo:
    columnas = ["sku", "nombre", "precio", "stock"]

    escritor = csv.DictWriter(archivo, fieldnames=columnas)
    escritor.writeheader()
    escritor.writerows(productos)

print("Archivo CSV creado correctamente:")
print(archivo_csv.resolve())
```

```python
def cargar_productos_csv():
    productos_cargados = []

    with open(archivo_csv, "r", encoding="utf-8") as archivo:
        lector = csv.DictReader(archivo)

        for fila in lector:
            producto = {
                "sku": fila["sku"],
                "nombre": fila["nombre"],
                "precio": float(fila["precio"]),
                "stock": int(fila["stock"])
            }

            productos_cargados.append(producto)

    return productos_cargados

productos_desde_csv = cargar_productos_csv()

valor_total_inventario = 0

print("Productos cargados desde CSV:")
for producto in productos_desde_csv:
    valor_producto = producto["precio"] * producto["stock"]
    valor_total_inventario += valor_producto

    print(
        f"{producto['sku']} | {producto['nombre']} | "
        f"${producto['precio']} | Stock: {producto['stock']} | "
        f"Valor: ${valor_producto:.2f}"
    )

print(f"\nValor total del inventario: ${valor_total_inventario:.2f}")
```

```python
#! Mini sistema de ventas con persistencia en JSON
archivo_sistema = CARPETA_DATOS / "sistema_ventas.json"

def cargar_sistema():
    if not archivo_sistema.exists():
        return {
            "productos": [],
            "ventas": []
        }

    with open(archivo_sistema, "r", encoding="utf-8") as archivo:
        return json.load(archivo)

def guardar_sistema(datos):
    with open(archivo_sistema, "w", encoding="utf-8") as archivo:
        json.dump(datos, archivo, indent=4, ensure_ascii=False)

def registrar_producto(sku, nombre, precio, stock):
    datos = cargar_sistema()

    for producto in datos["productos"]:
        if producto["sku"] == sku:
            print(f"No se registró el producto. El SKU {sku} ya existe.")
            return

    nuevo_producto = {
        "sku": sku,
        "nombre": nombre,
        "precio": precio,
        "stock": stock
    }

    datos["productos"].append(nuevo_producto)
    guardar_sistema(datos)

    print(f"Producto registrado: {nombre}")

def listar_productos():
    datos = cargar_sistema()

    if not datos["productos"]:
        print("No hay productos registrados.")
        return

    print("Productos disponibles:")
    for producto in datos["productos"]:
        print(
            f"{producto['sku']} | {producto['nombre']} | "
            f"${producto['precio']} | Stock: {producto['stock']}"
        )
```

```python
registrar_producto("P001", "Mouse inalámbrico", 249.90, 15)
registrar_producto("P002", "Teclado mecánico", 899.00, 8)
registrar_producto("P003", "Monitor 24 pulgadas", 3299.00, 4)
registrar_producto("P004", "Xbox", 9999.00, 5)

listar_productos()
```

```python
def buscar_producto(datos, sku):
    for producto in datos["productos"]:
        if producto["sku"] == sku:
            return producto

    return None

def registrar_venta(cliente, sku_producto, cantidad):
    datos = cargar_sistema()

    producto = buscar_producto(datos, sku_producto)

    if producto is None:
        print(f"No existe un producto con SKU {sku_producto}.")
        return

    if producto["stock"] < cantidad:
        print("No hay stock suficiente para realizar la venta.")
        return

    total = producto["precio"] * cantidad

    nueva_venta = {
        "id": len(datos["ventas"]) + 1,
        "fecha": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "cliente": cliente,
        "sku_producto": producto["sku"],
        "nombre_producto": producto["nombre"],
        "cantidad": cantidad,
        "precio_unitario": producto["precio"],
        "total": total
    }

    producto["stock"] -= cantidad
    datos["ventas"].append(nueva_venta)

    guardar_sistema(datos)

    print("Venta registrada correctamente.")
    print(f"Cliente: {cliente}")
    print(f"Producto: {producto['nombre']}")
    print(f"Cantidad: {cantidad}")
    print(f"Total: ${total:.2f}")

registrar_venta("Ana López", "P001", 2)
registrar_venta("Carlos Pérez", "P002", 1)

print("\nInventario después de las ventas:")
listar_productos()
```

```python
def listar_ventas():
    datos = cargar_sistema()

    if not datos["ventas"]:
        print("No hay ventas registradas.")
        return

    print("Ventas registradas:")
    for venta in datos["ventas"]:
        print(
            f"Venta {venta['id']} | {venta['fecha']} | "
            f"Cliente: {venta['cliente']} | "
            f"Producto: {venta['nombre_producto']} | "
            f"Cantidad: {venta['cantidad']} | "
            f"Total: ${venta['total']:.2f}"
        )

listar_ventas()
```

```python
def buscar_ventas_por_producto(sku_producto):
    datos = cargar_sistema()

    ventas_encontradas = []

    for venta in datos["ventas"]:
        if venta["sku_producto"] == sku_producto:
            ventas_encontradas.append(venta)

    if not ventas_encontradas:
        print(f"No se encontraron ventas para el producto {sku_producto}.")
        return

    print(f"Ventas encontradas para el producto {sku_producto}:")
    for venta in ventas_encontradas:
        print(
            f"Venta {venta['id']} | "
            f"Cliente: {venta['cliente']} | "
            f"Cantidad: {venta['cantidad']} | "
            f"Total: ${venta['total']:.2f}"
        )

buscar_ventas_por_producto("P001")
```

```python
def calcular_total_vendido_por_producto():
    datos = cargar_sistema()

    resumen = {}

    for venta in datos["ventas"]:
        sku = venta["sku_producto"]

        if sku not in resumen:
            resumen[sku] = {
                "nombre_producto": venta["nombre_producto"],
                "cantidad_total": 0,
                "importe_total": 0
            }

        resumen[sku]["cantidad_total"] += venta["cantidad"]
        resumen[sku]["importe_total"] += venta["total"]

    print("Total vendido por producto:")

    for sku, datos_producto in resumen.items():
        print(
            f"{sku} | {datos_producto['nombre_producto']} | "
            f"Unidades vendidas: {datos_producto['cantidad_total']} | "
            f"Importe total: ${datos_producto['importe_total']:.2f}"
        )

calcular_total_vendido_por_producto()
```

```python
import sqlite3

archivo_db = CARPETA_DATOS / "vista_previa_ventas.db"

conexion = sqlite3.connect(archivo_db)
cursor = conexion.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS productos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    sku TEXT NOT NULL UNIQUE,
    nombre TEXT NOT NULL,
    precio REAL NOT NULL,
    stock INTEGER NOT NULL
)
""")

cursor.execute("""
INSERT OR IGNORE INTO productos (sku, nombre, precio, stock)
VALUES
    ('P001', 'Mouse inalámbrico', 249.90, 15),
    ('P002', 'Teclado mecánico', 899.00, 8),
    ('P003', 'Monitor 24 pulgadas', 3299.00, 4)
""")

conexion.commit()

cursor.execute("SELECT sku, nombre, precio, stock FROM productos")

productos = cursor.fetchall()

print("Productos consultados desde SQLite:")
for producto in productos:
    print(producto)

conexion.close()
```
