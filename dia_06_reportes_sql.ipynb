```python
from pathlib import Path
import sqlite3
import pandas as pd

# Carpeta del Día 6.
CARPETA_DATOS = Path("datos_dia_06")
CARPETA_DATOS.mkdir(exist_ok=True)

# Base de datos del Día 6.
RUTA_DB = CARPETA_DATOS / "ventas_dia_06.db"

print("Carpeta de trabajo:")
print(CARPETA_DATOS.resolve())

print("\nBase de datos:")
print(RUTA_DB.resolve())
```

```python
conexion = sqlite3.connect(RUTA_DB)
cursor = conexion.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS productos (
    id_producto INTEGER PRIMARY KEY AUTOINCREMENT,
    sku TEXT NOT NULL UNIQUE,
    nombre TEXT NOT NULL,
    categoria TEXT NOT NULL,
    precio REAL NOT NULL CHECK(precio > 0),
    stock INTEGER NOT NULL CHECK(stock >= 0),
    activo INTEGER NOT NULL DEFAULT 1 CHECK(activo IN (0, 1))
)
""")

conexion.commit()
conexion.close()

print("Tabla productos creada correctamente.")
```

```python
from pathlib import Path
import sqlite3
import pandas as pd

CARPETA_DATOS = Path("datos_dia_06")
RUTA_DB = CARPETA_DATOS / "ventas_dia_06.db"

conexion = sqlite3.connect(RUTA_DB)

reporte_general = pd.read_sql_query("""
SELECT
    COUNT(*) AS total_productos_activos,
    SUM(stock) AS unidades_totales,
    ROUND(AVG(precio), 2) AS precio_promedio,
    MIN(precio) AS precio_minimo,
    MAX(precio) AS precio_maximo,
    ROUND(SUM(precio * stock), 2) AS valor_total_inventario
FROM productos
WHERE activo = 1
""", conexion)

valor_por_categoria = pd.read_sql_query("""
SELECT
    categoria,
    COUNT(*) AS cantidad_productos,
    SUM(stock) AS unidades_disponibles,
    ROUND(AVG(precio), 2) AS precio_promedio,
    ROUND(SUM(precio * stock), 2) AS valor_inventario
FROM productos
WHERE activo = 1
GROUP BY categoria
HAVING valor_inventario > 0
ORDER BY valor_inventario DESC
""", conexion)

productos_bajo_stock = pd.read_sql_query("""
SELECT
    sku,
    nombre,
    categoria,
    stock
FROM productos
WHERE activo = 1
  AND stock < 10
ORDER BY stock ASC
""", conexion)

conexion.close()

print("Reporte general de inventario")
display(reporte_general)

print("Valor de inventario por categoría")
display(valor_por_categoria)

print("Productos con bajo stock")
display(productos_bajo_stock)
```
